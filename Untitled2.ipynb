 Pre-processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('topinsta.csv')
print(data)
```

```python
# Basic overview
print("Dataset Info:")
print(data.info())
print("First few rows:")
print(data.head())
```

```python
# Check for missing values
print("Missing Values:")
print(data.isnull().sum())
```

```python
# Fill missing values in 'Category' and 'Audience Country' with 'Unknown'
data['Category'].fillna('Unknown', inplace=True)
data['Audience Country'].fillna('Unknown', inplace=True)
print("Missing values after filling:", data.isnull().sum())
```

```python
# Clean Numeric Columns with 'M' Notation
def convert_millions(value):
    if isinstance(value, str) and 'M' in value:
        return float(value.replace('M', '')) * 1_000_000
    return pd.to_numeric(value, errors='coerce')
```

```python
# Apply conversion to the specified columns
data['Followers'] = data['Followers'].apply(convert_millions)
data['Authentic Engagement'] = data['Authentic Engagement'].apply(convert_millions)
data['Engagement Avg.'] = data['Engagement Avg.'].apply(convert_millions)
```

```python
# Verify that the columns are now numeric
print(data[['Followers', 'Authentic Engagement', 'Engagement Avg.']].dtypes)
```

```python
# Display cleaned dataset's first few rows to confirm changes
data.head()
```

Follower and Engagement Analysis

```python
# Top 10 Accounts by Followers
top_followers = data.nlargest(10, 'Followers')[['Name', 'Followers']]
print("Top 10 Accounts by Followers:\n", top_followers)
```

```python
# Plot Follower Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Followers'], bins=30, kde=True)
plt.title('Follower Distribution')
plt.xlabel('Followers')
plt.ylabel('Frequency')

plt.show()
```

```python

# Correlation Analysis
correlation = data[['Followers', 'Authentic Engagement', 'Engagement Avg.']].corr()
print("Correlation between Followers, Authentic Engagement, and Engagement Avg.:\n", correlation)

# Heatmap of Correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
category_counts = data['Category'].value_counts()
print("Content Category Distribution:\n", category_counts)
```

```python
plt.figure(figsize=(10, 6))

# Get the top 10 categories
top_categories = category_counts.nlargest(10).index

# Filter category_counts to include only top categories
filtered_category_counts = category_counts[category_counts.index.isin(top_categories)]

# Plotting the filtered data
filtered_category_counts.plot(kind='bar')

plt.title('Content Category Distribution (Top 10)')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
top_countries = data['Audience Country'].value_counts().nlargest(10)
print("Top 10 Audience Countries:\n", top_countries)
```

```python
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Audience Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
avg_engagement_by_category = data.groupby('Category')['Engagement Avg.'].mean().sort_values(ascending=False)
print("Average Engagement by Category:\n", avg_engagement_by_category)
```

```python
plt.figure(figsize=(10, 6))

# Limit to top 12 categories
avg_engagement_by_category = avg_engagement_by_category.head(12)

avg_engagement_by_category.plot(kind='bar', color='salmon')
plt.title('Average Engagement per Category (Top 12)')
plt.xlabel('Category')
plt.ylabel('Average Engagement')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
